## 1. Setup & Path Detection

```python
import sys, os, json, io, math, glob, re, importlib.util
import torch, torch.nn.functional as F
import numpy as np

try:
    _nb = globals().get('__vsc_ipynb_file__') or __file__
    EXP_DIR = os.path.dirname(os.path.abspath(_nb))
except NameError:
    EXP_DIR = os.getcwd()

REPO_ROOT = os.path.abspath(os.path.join(EXP_DIR, '..', '..', '..'))
CHECKPOINT_DIR = os.path.join(EXP_DIR, 'checkpoint')
TOKENIZER_PATH = os.path.join(REPO_ROOT, 'data', 'tokenizers', 'fineweb_1024_bpe.model')
VAL_DATA_PATTERN = os.path.join(REPO_ROOT, 'data', 'datasets', 'fineweb10B_sp1024', 'fineweb_val_*.bin')
DEVICE = 'cuda' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu'

MODEL_PT  = os.environ.get('MODEL_PT',  os.path.join(EXP_DIR, 'checkpoint', 'model.pt'))
MODEL_PTZ = os.environ.get('MODEL_PTZ', os.path.join(EXP_DIR, 'checkpoint', 'model.int6.ptz'))

print(f'EXP_DIR   : {EXP_DIR}')
print(f'REPO_ROOT : {REPO_ROOT}')
print(f'DEVICE    : {DEVICE}')
print(f'model.pt  : {MODEL_PT}  — exists={os.path.exists(MODEL_PT)}')
print(f'int6.ptz  : {MODEL_PTZ} — exists={os.path.exists(MODEL_PTZ)}')
```

## 2. Load Model

```python
spec = importlib.util.spec_from_file_location('train_gpt', os.path.join(EXP_DIR, 'train_gpt.py'))
tg = importlib.util.module_from_spec(spec)
sys.path.insert(0, EXP_DIR)
spec.loader.exec_module(tg)
sys.path.pop(0)

import inspect
hp = tg.Hyperparameters()
valid_keys = set(inspect.signature(tg.GPT.__init__).parameters) - {'self'}
hp_dict = {k: getattr(hp, k) for k in valid_keys if hasattr(hp, k)}
model = tg.GPT(**hp_dict).eval()
print(f'Parameters: {sum(p.numel() for p in model.parameters()):,}')
print('Note: meta_sgd_{{qo,kv,up,down}} are registered but excluded from export.')
```

```python
# --- Float model ---
model_float = sd = None
if os.path.exists(MODEL_PT):
    sd = torch.load(MODEL_PT, map_location='cpu', weights_only=True)
    if isinstance(sd, dict) and 'model' in sd: sd = sd['model']
    m = tg.GPT(**hp_dict).eval()
    # meta_sgd keys are absent in exported pt; inject from fresh model
    fresh = tg.GPT(**hp_dict)
    for k in ('meta_sgd_qo','meta_sgd_kv','meta_sgd_up','meta_sgd_down'):
        if k not in sd and hasattr(fresh, k):
            sd[k] = getattr(fresh, k).detach().cpu().clone()
    m.load_state_dict(sd, strict=True)
    model_float = m.to(DEVICE)
    print(f'Loaded float model from {MODEL_PT}')
else:
    print('[skip] float model not found')
```

```python
# --- Int6 dequantized model ---
model_int6 = None
if os.path.exists(MODEL_PTZ):
    import lzma
    with open(MODEL_PTZ, 'rb') as f: blob = f.read()
    try:
        decompressed = lzma.decompress(blob)
    except Exception:
        import zlib; decompressed = zlib.decompress(blob)
    qs = torch.load(io.BytesIO(decompressed), map_location='cpu', weights_only=True)
    ref_sd = {k: v.cpu() for k, v in (sd or {}).items()}
    if not ref_sd and model_float:
        ref_sd = {k: v.cpu() for k, v in model_float.state_dict().items()}
    deq = tg.dequantize_mixed_int6(qs['w'], qs['m'], ref_sd)
    fresh = tg.GPT(**hp_dict)
    for k in ('meta_sgd_qo','meta_sgd_kv','meta_sgd_up','meta_sgd_down'):
        if k not in deq and hasattr(fresh, k):
            deq[k] = getattr(fresh, k).detach().cpu().clone()
    tg.CastedLinear._qat_enabled = True
    model_int6 = tg.GPT(**hp_dict).eval()
    model_int6.load_state_dict(deq, strict=True)
    model_int6 = model_int6.to(DEVICE)
    print(f'Loaded int6 model from {MODEL_PTZ}')
else:
    print('[skip] int6 model not found')
```

## 3. Compute val_bpb

```python
import sentencepiece as spm
sp = spm.SentencePieceProcessor(); sp.Load(TOKENIZER_PATH)

VAL_SHARDS = sorted(glob.glob(VAL_DATA_PATTERN))
assert VAL_SHARDS

def load_shard(path):
    hdr = np.fromfile(path, dtype='<i4', count=256)
    return np.fromfile(path, dtype='<u2', count=int(hdr[2]), offset=256*4)

val_tokens = np.concatenate([load_shard(s) for s in VAL_SHARDS])
print(f'Val tokens: {len(val_tokens):,}')

SEQ_LEN = hp.train_seq_len
BATCH_SEQ = 262144 // SEQ_LEN
LOG2E = math.log2(math.e)

def eval_bpb(m, toks, max_batches=4):
    m.eval(); total_loss = total_toks = 0; all_losses = []; all_ids = []; i = 0
    with torch.no_grad():
        for _ in range(max_batches):
            if i + SEQ_LEN*BATCH_SEQ+1 > len(toks): break
            chunk = toks[i:i+SEQ_LEN*BATCH_SEQ+1].astype(np.int64)
            x = torch.from_numpy(chunk[:-1]).reshape(BATCH_SEQ, SEQ_LEN).to(DEVICE)
            y = torch.from_numpy(chunk[1: ]).reshape(BATCH_SEQ, SEQ_LEN).to(DEVICE)
            logits = m.forward_logits(x)
            loss = F.cross_entropy(logits.reshape(-1, logits.size(-1)), y.reshape(-1), reduction='none')
            all_losses.append(loss.cpu()); all_ids.append(y.reshape(-1).cpu())
            total_loss += loss.sum().item(); total_toks += loss.numel(); i += SEQ_LEN*BATCH_SEQ
    ml = total_loss/total_toks
    return ml, ml*LOG2E, torch.cat(all_losses), torch.cat(all_ids)

if model_float:
    fl_loss, fl_bpb, fl_losses, fl_ids = eval_bpb(model_float, val_tokens)
    print(f'Float  val_loss={fl_loss:.4f}  val_bpb={fl_bpb:.4f}  (expected ~1.1377)')
if model_int6:
    q_loss, q_bpb, q_losses, q_ids = eval_bpb(model_int6, val_tokens)
    print(f'Int6   val_loss={q_loss:.4f}  val_bpb={q_bpb:.4f}  (expected ~1.1416)')
```

## 4. Text Generation

```python
def generate(m, prompt, max_new=150, temp=0.8, top_k=40):
    ids = sp.EncodeAsIds(prompt)
    x = torch.tensor(ids, dtype=torch.long, device=DEVICE).unsqueeze(0)
    m.eval()
    with torch.no_grad():
        for _ in range(max_new):
            logits = m.forward_logits(x)[:, -1, :] / temp
            if top_k:
                v, _ = torch.topk(logits, top_k)
                logits[logits < v[:, -1:]] = -float('inf')
            x = torch.cat([x, torch.multinomial(F.softmax(logits, -1), 1)], dim=1)
    return sp.DecodeIds(x[0].tolist())

active = model_float or model_int6
if active:
    for p in ['The history of artificial intelligence began',
              'In machine learning, test-time adaptation means']:
        print('='*60)
        print(generate(active, p))
        print()
```

```python
import matplotlib.pyplot as plt

if model_float:
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.hist(fl_losses.numpy(), bins=100, log=True, color='darkorange', alpha=0.8)
    ax.axvline(fl_losses.mean(), color='red', linestyle='--',
               label=f'mean={fl_losses.mean():.3f}')
    ax.set_title('exp106 (MetaSGD+CrossChunk+Δ) — Per-Token Loss Distribution')
    ax.set_xlabel('Cross-entropy'); ax.set_ylabel('Count (log)')
    ax.legend(); plt.tight_layout(); plt.show()
```

## 6. Per-Position Loss Curve

```python
if model_float:
    pos_acc = np.zeros(SEQ_LEN); pos_cnt = np.zeros(SEQ_LEN, dtype=np.int64)
    i = 0; model_float.eval()
    with torch.no_grad():
        for _ in range(4):
            if i + SEQ_LEN*BATCH_SEQ+1 > len(val_tokens): break
            chunk = val_tokens[i:i+SEQ_LEN*BATCH_SEQ+1].astype(np.int64)
            x = torch.from_numpy(chunk[:-1]).reshape(BATCH_SEQ, SEQ_LEN).to(DEVICE)
            y = torch.from_numpy(chunk[1: ]).reshape(BATCH_SEQ, SEQ_LEN).to(DEVICE)
            logits = model_float.forward_logits(x)
            loss = F.cross_entropy(logits.reshape(-1, logits.size(-1)),
                                   y.reshape(-1), reduction='none'
                                   ).reshape(BATCH_SEQ, SEQ_LEN).cpu().numpy()
            pos_acc += loss.sum(0); pos_cnt += BATCH_SEQ; i += SEQ_LEN*BATCH_SEQ
    pos_mean = pos_acc / pos_cnt
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.plot(pos_mean, lw=0.6, color='darkorange', label='raw')
    w = max(1, SEQ_LEN//64)
    ax.plot(np.convolve(pos_mean, np.ones(w)/w, 'same'), lw=2, color='red',
            alpha=0.7, label=f'smoothed (w={w})')
    ax.set_title('exp106 — Per-Position Mean Loss')
    ax.set_xlabel('Position'); ax.set_ylabel('Mean CE'); ax.legend()
    plt.tight_layout(); plt.show()
```

## 7. TTT Trajectory Visualization

Parse the partial int6 TTT log and the complete float TTT log to visualize bpb vs chunk.

```python
def parse_ttt_log(path):
    """Extract (chunk, bpb) pairs from a TTT progress log."""
    chunks, bpbs = [], []
    pattern = re.compile(r'chunk\s+(\d+)/(\d+).*bpb=([0-9.]+)')
    with open(path) as f:
        for line in f:
            m = pattern.search(line)
            if m:
                chunks.append(int(m.group(1)))
                bpbs.append(float(m.group(3)))
    return chunks, bpbs

float_log = os.path.join(EXP_DIR, 'ttt_from_checkpoint_float_qatoff.log')
int6_log  = os.path.join(EXP_DIR, 'ttt_int6_ep4_partial.log')

fig, ax = plt.subplots(figsize=(14, 5))

if os.path.exists(float_log):
    fc, fb = parse_ttt_log(float_log)
    ax.plot(fc, fb, lw=1.5, color='steelblue', label=f'Float TTT (final={fb[-1]:.4f})')
    ax.axhline(1.1377, color='steelblue', linestyle=':', alpha=0.5, label='Float baseline 1.1377')

if os.path.exists(int6_log):
    ic, ib = parse_ttt_log(int6_log)
    ax.plot(ic, ib, lw=1.5, color='darkorange', label=f'Int6 TTT partial (at {ic[-1]}/947: {ib[-1]:.4f})')
    ax.axhline(1.1416, color='darkorange', linestyle=':', alpha=0.5, label='Int6 baseline 1.1416')

ax.set_title('exp106 — TTT Trajectory (float complete, int6 partial 80%)')
ax.set_xlabel('Chunk index'); ax.set_ylabel('Running val_bpb')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```
